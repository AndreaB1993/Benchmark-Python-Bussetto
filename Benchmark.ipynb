## Trovare il numero di casi totali per ciascun continente (si escludono quelle location non appartenenti ad alcun continente)

```python
Subset1 = Covid19.loc[:,["continent", "new_cases"]].groupby("continent").sum() #Utilizzando la funzione .groupby() le location con continente nullo vengono escluse
Subset1.rename(columns={"new_cases" : "Tot_Casi"}, inplace=True)
Subset1.sort_values("Tot_Casi", ascending=False)
```

## Sviluppare una funzione che prenda in input il dataset e due continenti,  e che ne confronti i seguenti relativi descrittori statistici: valori minimo e massimo, media, e percentuale rispetto al numero dei casi totali nel mondo (calcolati anche sulle locazioni senza indicazione di continente);

```python
def Stat_casi(Covid19):
    Cont1=input("Inserire il nome di un continente: ").title() #L'utente da tastiera mette in input i due continenti
    Cont2=input("Inserire il nome di un continente: ").title() #La funzione .title() pone l'input nel formato già corretto del dataset


    S1 = Covid19.loc[Covid19.continent==Cont1].new_cases #Gli input vengono utilizzati come variabile per filtrare il dataset sul continente richiesto
    S2 = Covid19.loc[Covid19.continent==Cont2].new_cases #Le variabili necessarie per i vari calcoli si basano sulla colonna new_cases
    
    MaxC1=S1.max() #Si calcolano i massimi dei new_cases
    MaxC2=S2.max()
    
    MinC1=S1.min() #Si calcolano i minimi dei new_cases
    MinC2=S2.min()
    
    MeanC1=S1.mean() #Si calcola la media dei new_cases
    MeanC2=S2.mean()
    
    Tot_World_cases=Covid19.loc[~Covid19.continent.isnull()].new_cases.sum() 
    
    """
    Si calcolano i casi globali, escludendo i continenti nulli (trattandodi di aggregazioni continentali, sono da escludere nel computo complessivo)
    """
    
    World_Perc1=(S1.sum())/Tot_World_cases*100 #Si calcolano le percentuali sui casi globali
    World_Perc2=(S2.sum())/Tot_World_cases*100
    
    #Si struttura un piccolo dataframe per esporre in maniera chiara all'utente le informazioni richieste
    
    Stat=pd.DataFrame(
        {'Cont1': [MaxC1, MinC1, MeanC1, World_Perc1],
         'Cont2': [MaxC2, MinC2, MeanC2, World_Perc2]},
         index = ['Massimo casi Giorn.', 'Minimo casi Giorn.Giorn.', 'Media casi Giorn.', '% su Mondo'])
    Stat.columns = [Cont1, Cont2]
    
    pd.options.display.float_format = '{:,.2f}'.format
    
    return Stat
```

```python
Stat_casi(Covid19)
```

## Si chiede poi di effettuare lo stesso tipo di analisi – anche in questo caso sviluppando una funzione ad hoc – per il numero di vaccinazioni totali per ogni continente;

```python
def Stat_vaccini(Covid19):
    Cont1=input("Inserire il nome di un continente: ").title() #L'utente da tastiera mette in input i due continenti
    Cont2=input("Inserire il nome di un continente: ").title() #La funzione .title() pone l'input nel formato già corretto del dataset

    S1 = Covid19.loc[Covid19.continent==Cont1].new_vaccinations #Gli input vengono utilizzati come variabile per filtrare il dataset sul continente richiesto
    S2 = Covid19.loc[Covid19.continent==Cont2].new_vaccinations #Le variabili necessarie per i vari calcoli si basano sulla colonna new_vaccination
    
    MaxV1=S1.max() #Si calcolano i massimi delle new_vaccination
    MaxV2=S2.max()
    
    MinV1=S1.min() #Si calcolano i minimi delle new_vaccination
    MinV2=S2.min()
    
    MeanV1=S1.mean() #Si calcolano la media delle new_vaccination
    MeanV2=S2.mean()
    
    Tot_World_Vax=Covid19.loc[~Covid19.continent.isnull()].new_vaccinations.sum()
    
    """
    Si calcolano le vacinazioni globali, escludendo i continenti nulli (trattandodi di aggregazioni continentali, sono da escludere nel computo complessivo)
    """
    World_Perc1=(S1.sum())/Tot_World_Vax*100 #Si calcolano le percentuali sulle vaccinazioni globali
    World_Perc2=(S2.sum())/Tot_World_Vax*100
    
    #Si struttura un piccolo dataframe per esporre in maniera chiara all'utente le informazioni richieste
    
    Stat=pd.DataFrame(
        {'Cont1': [MaxV1, MinV1, MeanV1, World_Perc1],
         'Cont2': [MaxV2, MinV2, MeanV2, World_Perc2]},
         index = ['Massimo Vax Giorn.', 'Minimo Vax Giorn.', 'Media Vax Giorn.', '% su Mondo'])
    Stat.columns = [Cont1, Cont2]
    
    pd.options.display.float_format = '{:,.2f}'.format
    
    return Stat
```

```python
Stat_vaccini(Covid19)
```

## Alla fine, basandosi sui calcoli fatti, il committente chiede di stilare un breve (tre o quattro righe) paragrafo testuale riassuntivo sulle statistiche di casi e vaccinazioni, che si concentri solo sulle differenze esistenti tra Europa, Sud America e Oceania.

```python
def Stat_EU_SAM_OC(Covid19):
    
    S1 = Covid19.loc[Covid19.continent=="Europe"] #Gli input vengono utilizzati come variabile per filtrare il dataset sul continente richiesto
    S2 = Covid19.loc[Covid19.continent=="South America"] #Le variabili necessarie per i vari calcoli si basano sulla colonna new_vaccination
    S3 = Covid19.loc[Covid19.continent=="Oceania"]
    

    TotC1=S1.new_cases.sum() #Si calcolano i massimi delle new_vaccination
    TotC2=S2.new_cases.sum()
    TotC3=S3.new_cases.sum()
    
    MaxC1=S1.new_cases.max() #Si calcolano i massimi delle new_vaccination
    MaxC2=S2.new_cases.max()
    MaxC3=S3.new_cases.max()

    MinC1=S1.new_cases.min() #Si calcolano i massimi delle new_vaccination
    MinC2=S2.new_cases.min()
    MinC3=S3.new_cases.min()
    
    MeanC1=S1.new_cases.mean() #Si calcolano la media delle new_vaccination
    MeanC2=S2.new_cases.mean()
    MeanC3=S3.new_cases.mean()
    
    MaxV1=S1.new_vaccinations.max() #Si calcolano i massimi delle new_vaccination
    MaxV2=S2.new_vaccinations.max()
    MaxV3=S3.new_vaccinations.max()

    MinV1=S1.new_vaccinations.min() #Si calcolano i massimi delle new_vaccination
    MinV2=S2.new_vaccinations.min()
    MinV3=S3.new_vaccinations.min()
    
    MeanV1=S1.new_vaccinations.mean() #Si calcolano la media delle new_vaccination
    MeanV2=S2.new_vaccinations.mean()
    MeanV3=S3.new_vaccinations.mean()
    
    Tot_World_cases=Covid19.loc[~Covid19.continent.isnull()].new_cases.sum()
    Tot_World_Vax=Covid19.loc[~Covid19.continent.isnull()].new_vaccinations.sum()
    """
    Si calcolano le vacinazioni globali, escludendo i continenti nulli (trattandodi di aggregazioni continentali, sono da escludere nel computo complessivo)
    """
    
    World_Perc1_Cases=(S1.new_cases.sum())/Tot_World_cases*100 #Si calcolano le percentuali sui casi globali
    World_Perc2_Cases=(S2.new_cases.sum())/Tot_World_cases*100
    World_Perc3_Cases=(S3.new_cases.sum())/Tot_World_cases*100
    
    World_Perc1_Vax=(S1.new_vaccinations.sum())/Tot_World_Vax*100 #Si calcolano le percentuali sulle vaccinazioni globali
    World_Perc2_Vax=(S2.new_vaccinations.sum())/Tot_World_Vax*100
    World_Perc3_Vax=(S3.new_vaccinations.sum())/Tot_World_Vax*100
    
    print(f"L'Europa ha avuto complessivamente {TotC1:,.2f} casi, corrispondenti al {World_Perc1_Cases:,.2f}% dei casi totali.")
    print(f"Con un massimo di {MaxC1} casi giornalieri ed una media di {MeanV1:,.2f} casi.")
    print(f"Le vaccinazioni eseguite in Europa sono state il {World_Perc1_Vax:,.2f}% del totale mondiale.")
    print(f"Con un massimo di {MaxV1:,.2f} vaccinazioni giornaliere.")
    
    print("\n")
    
    print(f"Il Sud Amrica ha avuto complessivamente {TotC2:,.2f} casi, corrispondenti al {World_Perc2_Cases:,.2f}% dei casi totali.")
    print(f"Con un massimo di {MaxC2:,.2f} casi giornalieri ed una media di {MeanV2:,.2f} casi.")
    print(f"Le vaccinazioni eseguite in Europa sono state il {World_Perc2_Vax:,.2f}% del totale mondiale.")
    print(f"Con un massimo di {MaxV2:,.2f} vaccinazioni giornaliere.") 
    
    print("\n")
    
    print(f"L'Oceania ha avuto complessivamente {TotC3:,.2f} casi, corrispondenti al {World_Perc3_Cases}% dei casi totali.")
    print(f"Con un massimo di {MaxC3:,.2f} casi giornalieri ed una media di {MeanV3:,.2f} casi.")
    print(f"Le vaccinazioni eseguite in Europa sono state il {World_Perc3_Vax:,.2f}% del totale mondiale.")
    print(f"Con un massimo di {MaxV3:,.2f} vaccinazioni giornaliere.")
    
    return
```
